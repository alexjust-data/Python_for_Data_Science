```python
#pero tweepy tiene una "classe" implementada para no preocuparnos de la paginación tweepy.Cursor()
#tweepy.Cursor() -> podré obtener los items que yo quiera, sin tope de paginacion de lo que quiera, followers, likes, etc
#tweepy.Cursor(api.followers,) -> li 1ro que recibe es cuál es el método que no estoy llamando
#tweepy.Cursor(api.followers,) -> le paso el método "api.followers()", no la llamada al método "api.followers"
#tweepy.Cursor(api.followers, screen_name = "nike") le paso en nombre del usuario que estamos mirando
#tweepy.Cursor(api.followers, screen_name = "nike").items(100) -> le digo el limite de items que quiero obtener
#ahora si i sí, todo esto lo tengo que iterar
for user in tweepy.Cursor(api.followers, screen_name = "nike").items(1): #le digo sólo de un usuario
  print (json.dumps(user._json, indent = 4))
```

```python
#VAMOS A VER A QUIEN SIGUE NIKE 
#obtener friends/followeers de un usuario utilizando Cursor()
for user in tweepy.Cursor(api.friends, screen_name = "nike").items(1): #le digo sólo de un usuario
  print (json.dumps(user._json, indent = 4))
```

```python
#OBTENER UN TIMELIME de un usuario, son todos los tweets que ha hecho un usuario en su perfil
#tweet_mode= longitud del parámetro que voy a obtener. (máx twitter 260 carácters)
#tweet_mode= "extended" -> si quiero todo
for tweet in tweepy.Cursor(api.user_timeline, 
                          screen_name = "nike", 
                          tweet_mode= "extended").items(1): #le digo sólo de un usuario
  print (json.dumps(tweet._json, indent = 4))
```

```python
#BUSCAR TWEETS
#hay muchos parámetros para la búsqueda, uno de ellos el "q"
# q = "nos permite búsqueda por texto" 

for tweet in tweepy.Cursor(api.search, 
                           q = "mundial de clubes",
                           #le digo sólo de un usuario
                           tweet_mode= "extended").items(1): 
    print (json.dumps(tweet._json, indent = 4))
```
